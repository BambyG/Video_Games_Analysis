```python
df['User_Score'] =  pd.to_numeric(df['User_Score'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
med_criticscore = df['Critic_Score'].median()
med_criticcount = df['Critic_Count'].median()
med_userscore = df['User_Score'].median()
med_userscount = df['User_Count'].median()

df['Critic_Score']= df['Critic_Score'].fillna(med_criticscore)
df['Critic_Count']= df['Critic_Count'].fillna(med_criticcount)
df['User_Score'] = df['User_Score'].fillna(med_userscore)
df['User_Count'] = df['User_Count'].fillna(med_userscount)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['Developer','Rating'],axis=1)
```

```python
df = df.dropna()
```

```python
from bokeh.plotting import figure, output_file, show

# prepare some data
x = df['EU_Sales']
y = df['NA_Sales']

# output to static HTML file
output_file("lines.html")

# create a new plot with a title and axis labels
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.circle(x, y, legend="Temp.", line_width=2)

# show the results
show(p)
```
